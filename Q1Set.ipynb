**Q3:** Given an integer A, and two integers B and C in base A. Perform the addition of B and C in base A and return the result in base A.

-The first argument is an integer A, which is the base  of the other two numbers.
-The second argument is the first number B in base A.
-The third argument is the second number C in base A.
-Return the addition of B and C in base A.

Sample Run:
Input:
A = 3
B = 11
C = 10

Output:
21

Explaination:
11 is the base 3 representation of decimal number 4.
10 is the base 3 representation of decimal number 3.
And sum of 3 and 4 in decimal representation is 7, which is 21 in base 3 representation.




```python
class Solution:
    # @param A : integer
    # @param B : integer
    # @param C : integer
    # @return an integer
    def AnyBaseAddition(self, A, B, C):
        sum1 = 0
        c = 0
        p = 1
        while B > 0 or C > 0 or c > 0:
            d1 = B % 10
            d2 = C % 10
            d = d1 + d2 + c
            c = d // A
            d = d % A
            sum1 += d * p
            p = p * 10
            B = B // 10
            C = C // 10
        return sum1
```
